```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
count = purchase_data["SN"].nunique()
totalplayer = pd.DataFrame({"Total Players":[count]})
totalplayer
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = purchase_data["Item ID"].nunique()
average = purchase_data["Price"].mean()
purchase = purchase_data["Purchase ID"].count()
total = purchase_data["Price"].sum()
summarytable = pd.DataFrame({"Number of Unique Items":[items],
                             "Average Price":[average],
                              "Number of Purchases":[purchase],
                              "Total Revenue":[total]})
summarytable
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
gendergroup = purchase_data["Gender"].value_counts()
gendergroup_df = pd.DataFrame(gendergroup)
gendergroup_df = gendergroup_df.rename(columns={"Gender": "Total Count"})
percent = (gendergroup/purchase)*100

gendergroup_df["% of Total"]=percent
gendergroup_df.head()

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gendercounts = purchase_data["Gender"].value_counts()
gendercounts.head()
```

```python
groupedgender_df = purchase_data.groupby(['Gender'])
print(groupedgender_df)
groupedgender_df.head()
```

## Age Demographics

```python
totalpricegroup = groupedgender_df["Price"].sum()
averagegroup = groupedgender_df["Price"].mean()
normalizedgroup = totalpricegroup/gendercounts

summarytable1=pd.DataFrame({"Purchase Count":gendercounts,
                            "Total Purchase Price":totalpricegroup,
                            "Average Purchase Price":averagegroup,
                            "Normalized":normalizedgroup})
summarytable1.head()
```

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] =pd.cut(purchase_data["Age"], age_bins, labels=group_names)
agegroup = purchase_data.groupby("Age Group")

percentagebyage = agegroup["SN"].nunique()/(agegroup["SN"].nunique()).sum() * 100
totalcount = agegroup["SN"].nunique()
summarytable2=pd.DataFrame({"Total Count":totalcount,
   "Percent":percentagebyage})
summarytable2
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Group"] =pd.cut(purchase_data["Age"], age_bins, labels=group_names)
agegroup = purchase_data.groupby("Age Group")

agecounts = purchase_data["Age Group"].value_counts()
totalagegroup = agegroup["Price"].sum()
averageagegroup = agegroup["Price"].mean()
normalizedagegroup = totalagegroup/agecounts

summarytable4=pd.DataFrame({"Total Counts":agecounts,
                            "Total Purchase Price":totalagegroup,
                            "Average Purchase Price":averageagegroup,
                            "Normalize Purchase Price":normalizedagegroup
                            })
summarytable4.head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
groupedsn_df = purchase_data.groupby(['SN'])
print(groupedsn_df)

totalsncount = groupedsn_df["SN"].count()
totalpricesn = groupedsn_df["Price"].sum()
averagepricesn= groupedsn_df["Price"].mean()



summarytable5=pd.DataFrame({"Total Count":totalsncount,
                            "Total Purchase":totalpricesn,
                            "Average Purchase Price":averagepricesn})
summarytable5 = summarytable5.sort_values(["Total Purchase"], ascending=False)
summarytable5.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
groupeditem_df = purchase_data.groupby(['Item ID'])
print(groupeditem_df)

totalitemcount = groupeditem_df["Item ID"].count()
totalitemprice = groupeditem_df["Price"].sum()
itemprice = totalitemprice/totalitemcount



summarytable6=pd.DataFrame({"Purchase Count":totalitemcount,
                            "Total Purchase":totalitemprice,
                            "Item Price":itemprice})

summarytable6 = summarytable6.sort_values(["Purchase Count"], ascending=False)

summarytable6.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
groupeditem_df = purchase_data.groupby(['Item ID'])
print(groupeditem_df)

totalitemcount = groupeditem_df["Item ID"].count()
totalitemprice = groupeditem_df["Price"].sum()
itemprice = totalitemprice/totalitemcount



summarytable6=pd.DataFrame({"Purchase Count":totalitemcount,
                            "Total Purchase":totalitemprice,
                            "Item Price":itemprice})

summarytable6 = summarytable6.sort_values(["Total Purchase"], ascending=False)

summarytable6.head()
```
